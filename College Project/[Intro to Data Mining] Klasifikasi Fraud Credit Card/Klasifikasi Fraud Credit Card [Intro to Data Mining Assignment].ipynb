# **Narasi**
**PDM Paylater** adalah perusahaan fiktif penyedia layanan Buy Now Pay Later (BNPL) yang utamanya mendapatkan pendapatan dengan memberikan pinjaman uang. **Tugas Anda** adalah membantu PDM Paylater dalam mengatasi tantangan ini. ## 2 - Exploratory Data Analysis (EDA)
Tahapan ini dilakukan untuk mendapatkan insight yang bermanfaat dari data. Pada tahapan ini Anda harus membuat setidaknya 5 pertanyaan tambahan yang akan dijawab berdasarkan hasil eksplorasi dan visualisasi selain 3 pertanyaan wajib

#### **2.1 - Bagaimana sebaran proporsi pada label target (`FLAG`)?**
---

Visualisasi yang dapat membantu menjawab pertanyaan ini adalah *bar chart* atau *pie chart*.

##### <b><span style='color:#F1C40F'> Visualisasi |</span> <em>Bar Chart</em> FLAG</b>

```python
# Mengatur gaya plot
sns.set_style('whitegrid')

# Membuat figure dan mengatur background
plt.figure(figsize=(10, 6), facecolor='#b5b5b5')

# Membuat barplot dengan warna gradien
unique_flags = df_tr['FLAG'].unique()
palette = sns.color_palette('viridis', len(unique_flags))
barplot = sns.countplot(x='FLAG', data=df_tr, palette=palette)

# Menambahkan warna gradien pada setiap bar
for i, bar in enumerate(barplot.patches):
    bar.set_facecolor(palette[i % len(palette)])
    # Menambahkan jumlah di tengah bar
    height = bar.get_height()
    barplot.text(bar.get_x() + bar.get_width() / 2.0, height / 2.0, int(height), ha='center', va='center', fontproperties=font_bold, color='white')

plt.title('Distribusi FLAG', fontproperties=font_title)
plt.xlabel('FLAG', fontproperties=font)
plt.ylabel('Count', fontproperties=font)

# Menampilkan plot
plt.show()
```

##### <b><span style='color:#F1C40F'> Visualisasi |</span> <em>Pie Chart</em> FLAG</b>

```python
# Mengatur gaya plot
sns.set_style('whitegrid')

# Membuat figure dan mengatur background
plt.figure(figsize=(10, 6), facecolor='#b5b5b5')

# Mengatur palet warna untuk setiap kategori FLAG
palette = {0: '#1f77b4', 1: '#ff7f0e'}

# Pie chart
df_tr['FLAG'].value_counts().plot.pie(autopct='%1.1f%%', textprops={'color': 'white'})

plt.title('Distribusi FLAG', fontproperties=font_title)
plt.legend(['Pinjaman Disetujui', 'Pinjaman Ditolak'], loc='upper right', prop=font)

# Mengatur label sumbu y menjadi kosong
plt.ylabel('')

# Menampilkan plot
plt.show()
```

<div class="alert alert-block alert-info">
<b>Jawab:</b> Berdasarkan visualisasi yang telah dibuat, diperoleh hasil bahwa ternyata distribusi FLAG bersifat <em>imbalance</em> karena proporsinya tidak serupa. FLAG 0 (seseorang diberikan pinjaman) pada data berjumlah 91,9% sedangkan sisanya adalah FLAG 1 (seseorang tidak diberikan pinjaman). Kasus <em>imbalance</em> ini mungkin akan menyebabkan terpengaruhnya model prediktif yang akan dibuat.</div>

Sesuai dengan yang telah diutarakan dalam situs [OJK](https://sikapiuangmu.ojk.go.id/FrontEnd/CMS/Article/10423), terdapat pengelompokan pelanggan berdasarkan telat bayar mereka--disebut kolektibilitas.

1.      Kredit Lancar atau Kol 1: Kredit yang memuaskan dimana kamu mampu menyelesaikan segala kewajibanmu seperti angsuran, pokok utang, dan bunga tanpa ada cela.

2.      Kredit Dalam Perhatian Khusus (DPK) atau Kol 2:  Terdapat tunggakan selama 1-2 bulan yang biasanya disebabkan karena keterlambatan pembayaran.

3.      Kredit Tidak Lancar atau Kol 3:   Terdapat tunggakan selama kurun 3-4 bulan. Pendekatan yang dilakukan kepada nasabah pun tidak membuahkan hasil.

4.      Kredit Diragukan atau Kol 4:   Kredit tidak lancar yang telah jatuh tempo tapi belum juga diselesaikan oleh Debitur lebih dari 5-6 bulan.

5.      Kredit Macet atau Kol 5:   Kredit tidak lancar yang tertunggak lebih dari 6 bulan dan telah diusahakan untuk diaktifkan kembali tapi tetap tidak membuahkan hasil.

```python
# If AVG_DELAY_DAYS is negative or 0, add Kol column value with 0
# If AVG_DELAY_DAYS is 0 < x <= 60, add Kol column value with 1
# If AVG_DELAY_DAYS is 60 < x <= 120, add Kol column value with 2
# If AVG_DELAY_DAYS is 120 < x <= 180, add Kol column value with 3
# If AVG_DELAY_DAYS is 180 < x, add Kol column value with 4

df_tr['Kol'] = 0
df_tr.loc[(df_tr['AVG_DELAY_DAYS'] > 0) & (df_tr['AVG_DELAY_DAYS'] <= 60), 'Kol'] = 1
df_tr.loc[(df_tr['AVG_DELAY_DAYS'] > 60) & (df_tr['AVG_DELAY_DAYS'] <= 120), 'Kol'] = 2
df_tr.loc[(df_tr['AVG_DELAY_DAYS'] > 120) & (df_tr['AVG_DELAY_DAYS'] <= 180), 'Kol'] = 3
df_tr.loc[df_tr['AVG_DELAY_DAYS'] > 180, 'Kol'] = 4

# Do the same for df_te on Kol

df_te['Kol'] = 0
df_te.loc[(df_te['AVG_DELAY_DAYS'] > 0) & (df_te['AVG_DELAY_DAYS'] <= 60), 'Kol'] = 1
df_te.loc[(df_te['AVG_DELAY_DAYS'] > 60) & (df_te['AVG_DELAY_DAYS'] <= 120), 'Kol'] = 2
df_te.loc[(df_te['AVG_DELAY_DAYS'] > 120) & (df_te['AVG_DELAY_DAYS'] <= 180), 'Kol'] = 3
df_te.loc[df_te['AVG_DELAY_DAYS'] > 180, 'Kol'] = 4
```

### Pemodelan
- Pilihlah metrik evaluasi yang Anda gunakan untuk menilai performa model (tahapan pemodelan serta pemilihan model terbaik) dan berikan penjelasan menyeluruh tentang alasan Anda memilih metrik tersebut.
- Pilihlah model terbaik dengan membandingkan beberapa model klasifikasi (**minimal 7 model yang berbeda**) dan berikan penjelasan mengapa model tersebut dipilih sebagai yang terbaik.

**Metrik evaluasi** yang digunakan: $F_{\beta}$-score, *recall*, dan AUC.
<br>
Karena $F_\beta$-score merupakan rataan harmonik dari presisi dan *recall*. Parameter $\beta$ merepresentasikan rasio kepentingan *recall* dibandingkan presisi. Nilai $\beta > 1$ mengindikasikan bahwa *recall* lebih dipentingkan daripada presisi. Dalam kasus ini, *recall* memang akan lebih dipentingkan daripada presisi karena *recall* merujuk pada sensitivitas. Kesalahan memprediksi seseorang gagal bayar dianggap lebih merugikan daripada kesalahan memprediksi seseorang berhasil bayar.

*Recall* digunakan karena mengukur kemampuan model untuk menangkap semua kasus positif (gagal bayar). Dalam konteks kredit, kesalahan memprediksi nasabah yang sebenarnya gagal bayar sebagai berhasil bayar (*false negative*) memiliki konsekuensi finansial yang lebih signifikan dibandingkan kesalahan sebaliknya. Oleh karena itu, meningkatkan *recall* membantu memastikan bahwa lebih banyak kasus gagal bayar teridentifikasi, mengurangi potensi kerugian.

AUC (Area Under the Curve) juga digunakan karena memberikan gambaran tentang kemampuan model dalam membedakan antara kelas positif dan negatif pada berbagai threshold. AUC adalah metrik yang kuat untuk mengukur performa model pada dataset yang tidak seimbang, karena tidak dipengaruhi oleh distribusi kelas. Dengan AUC, kita bisa memahami seberapa baik model dalam membedakan antara pengguna PayLater yang gagal bayar dan yang berhasil bayar, tanpa bias terhadap proporsi kelas yang berbeda. AUC yang tinggi menunjukkan bahwa model memiliki performa yang baik secara keseluruhan.

Mempertimbangkan adanya ketidakseimbangan pada data pula, metrik evaluasi akurasi tidak dapat digunakan karena hasilnya menjadi bias dengan jumlah variabel target. Secara singkat, akan dipilih $\beta = 2$ karena *recall* akan dipentingkan dua kali lebih penting daripada presisi.
PS: Saat ini praktikan belum dapat memberikan justifikasi lebih lanjut terkait dipilihnya $\beta = 2$.

```python
# SMOTETOMEK
smotetomek = SMOTETomek(random_state=0)
X_train_scaled, y_train = smotetomek.fit_resample(X_train_scaled, y_train)
```

```python
rand = 0
models = {
    'Random Forest': RandomForestClassifier(random_state=rand),
    'SVM': SVC(random_state=rand),
    'Naive-Bayes': GaussianNB(),
    'kNN': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(random_state=rand),
    'Gradient Boosting': GradientBoostingClassifier(random_state=rand),
    'XGBoost': XGBClassifier(random_state=rand),
    'MLP': MLPClassifier(random_state=rand),
    'Logistic Regression': LogisticRegression(random_state=rand),
    'AdaBoost': AdaBoostClassifier(random_state=rand)
}

# Initialize results dataframe
results = pd.DataFrame(columns=['Model', 'F-beta', 'AUC', 'Recall'])

# Loop over models
for name, model in models.items():
    pipeline = Pipeline([
        ('classifier', model)
    ])
    
    # Fit model
    pipeline.fit(X_train_scaled, y_train)
    
    # Predict and evaluate
    y_pred = pipeline.predict(X_val_scaled)
    # Check if model has predict_proba
    if hasattr(pipeline.named_steps['classifier'], 'predict_proba'):
        y_proba = pipeline.predict_proba(X_val_scaled)[:, 1]
        auc = roc_auc_score(y_val, y_proba)
    else:
        auc = None  # Set AUC to None if predict_proba is not available
    
    fbeta = fbeta_score(y_val, y_pred, beta=2)
    recall = recall_score(y_val, y_pred)
    
    results = pd.concat([results, pd.DataFrame({    
        'Model': [name],
        'F-beta': [fbeta],
        'AUC': [auc],
        'Recall': [recall]
    })], ignore_index=True)

# Display results
print(results)
```

Model yang paling baik adalah regresi logistik karena dari ketiga evaluasi yang digunakan, dua di antaranya memiliki evaluasi paling baik.

### Feature Importance

```python
from sklearn.inspection import permutation_importance
import shap
```

```python
def plot_fi(model, X = X_val_scaled, y = y_val, rep = 10):
    perm = permutation_importance(model, X, y, n_repeats=10, random_state=0, n_jobs=-1)

    perm2 = pd.DataFrame({'feature': X.columns.tolist()*rep, 'importance': perm["importances"].transpose().reshape(-1)})
    perm2["importance"] = perm2.importance/perm2.importance.sum()*rep
    urut = perm2.groupby("feature").mean().sort_values("importance", ascending=False)

    plt.figure(figsize=(8, 6))
    fig, axs = plt.subplots(1, 2, figsize=(25, 15))

    sns.barplot(x='importance', y='feature', data=perm2, order=urut.index, ax=axs[0])
    sns.stripplot(x='importance', y='feature', data=perm2, order = urut.index, s=5, ax=axs[1])
    plt.show()

    return urut
```

```python
plot_fi(model = best, X = X_val_scaled, y = y_val)
```
